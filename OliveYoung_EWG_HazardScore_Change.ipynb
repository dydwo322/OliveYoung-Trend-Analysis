```python
comp = pd.read_csv('title_id_compo_skin_up200.csv' ,encoding ='cp949' )
score = pd.read_csv('ing_final (2).csv' )
```

```python
import numpy as np

dic_score = {
    'product_id' : [],
    'title' : [],
    'score' : []
}


for i in tqdm(range(len(comp))) :

  dic_score['title'].append(comp['title'][i])
  dic_score['product_id'].append(comp['product_id'][i])
  score1 = []
  list1 = []  
  list1.extend(comp['components'][i].split(','))

  for j in range(len(list1)) :

    score1.extend(list( score[  (score['표준 성분명'] == list1[j]) | (score['구명칭'] == list1[j]) ]['avg_score'] ) ) 
  
  dic_score['score'].append(np.mean(score1))
    

score_ = pd.DataFrame(dic_score)

score_.to_csv('등급점수2.csv', index = False, encoding = 'utf-8-sig')

```
