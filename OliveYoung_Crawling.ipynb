```python
from requests import request


def getTotalReviewCount(gdid):
    url='https://www.oliveyoung.co.kr/store/goods/getGoodsGtmInfoAjaxNew.do?'
    params={
        'goodsNoArrStr':gdid,
        'itemNoArrStr':'001',
        'positionArrStr':'1',
        'giftCardYn':'N'
    }
    resp = requests.get(url,params=params)
    jsn=json.loads(resp.text)[0]
    return int(jsn['prmumGdasTotCnt'])
```
